```python
%%writefile course1_app.py

import streamlit as st
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# ---------------- Load Dataset ----------------
@st.cache_data
def load_data():
    return pd.read_csv("courses.csv")

data = load_data()

# Combine dataset features
data['combined'] = data['Interests'] + " " + data['Past_Courses'] + " " + data['Skills']

# ---------------- TF-IDF ----------------
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(data['combined'])

# ---------------- UI ----------------
st.set_page_config(page_title="Course Recommendation System", layout="wide")
st.title("🎓 Course Recommendation System")

st.markdown("### 🔹 Student Details (Select Options)")

# Dropdown values from dataset
interest_list = sorted(data['Interests'].unique())
course_list = sorted(data['Past_Courses'].unique())

# Skill list (manual – realistic)
skill_list = [
    "Python", "Java", "C++", "SQL", "HTML", "CSS",
    "JavaScript", "Machine Learning", "Data Analysis",
    "Cloud", "Linux", "Networking"
]

# ----------- Selection Inputs -----------
interest = st.selectbox("1️⃣ Area of Interest", interest_list)
past_course = st.selectbox("2️⃣ Previously Completed Course", course_list)
skills = st.multiselect("3️⃣ Skills", skill_list)

# ----------- Prediction -----------
if st.button("🔍 Predict Recommended Course"):
    if skills:
        student_profile = interest + " " + past_course + " " + " ".join(skills)

        input_vec = tfidf.transform([student_profile])
        similarity = cosine_similarity(input_vec, tfidf_matrix).flatten()
        top_index = similarity.argmax()

        st.success("✅ Recommended Course")

        st.markdown(
            f"""
            <div style="background-color:52C4AD;padding:15px;border-radius:10px;">
            <h3>🎯 {data.iloc[top_index]['Recommended_Course']}</h3>
            <p>
            <b>Interest:</b> {interest}<br>
            <b>Matched Skills:</b> {data.iloc[top_index]['Skills']}
            </p>
            </div>
            """,
            unsafe_allow_html=True
        )
    else:
        st.warning("⚠️ Please select at least one skill")

# ---------------- Trending Courses ----------------
st.markdown("---")
st.subheader("🔥 Trending Courses")
st.write("• Artificial Intelligence")
st.write("• Full Stack Development")
st.write("• Data Analytics")
st.write("• Cloud Computing")
st.write("• Cyber Security")
```
